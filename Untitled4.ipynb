```python
%%writefile core/gestor.py
import json
from datetime import datetime

def cargar_datos():
    with open("data/datos.json", "r") as archivo:
        return json.load(archivo)

def guardar_datos(datos):
    with open("data/datos.json", "w") as archivo:
        json.dump(datos, archivo, indent=4)

def calcular_balance():
    datos = cargar_datos()
    total_ingresos = sum(i["monto"] for i in datos["ingresos"])
    total_gastos = sum(g["monto"] for g in datos["gastos"])
    return total_ingresos - total_gastos

def validar_gasto(descripcion, monto, fecha, categoria):
    if not descripcion.strip():
        print(" Error: Ingrese una descripción válida")
        return False
    if monto <= 0:
        print(" Error: Ingrese un monto válido (mayor que 0)")
        return False

    try:
        datetime.strptime(fecha, "%d/%m/%Y")
    except ValueError:
        print(" Error: Fecha inválida. Use formato dd/mm/aaaa")
        return False


    if not categoria.strip():
        print(" Error: Seleccione una categoría")
        return False


    if monto > 10000:
        print(" Error: Límite diario excedido (máx. 10,000)")
        return False

    return True

def listar_gastos():
    datos = cargar_datos()
    if not datos["gastos"]:
        print(" No hay gastos registrados")
    else:
        for gasto in datos["gastos"]:
            print(gasto)

def crear_gasto(descripcion, monto, fecha, categoria):
    datos = cargar_datos()
    balance = calcular_balance()

    if not validar_gasto(descripcion, monto, fecha, categoria):
        return

    if monto > balance:
        print(" Error: No tienes suficiente saldo para registrar este gasto")
        return

    nuevo_id = len(datos["gastos"]) + 1
    nuevo_gasto = {
        "id": nuevo_id,
        "descripcion": descripcion,
        "monto": monto,
        "fecha": fecha,
        "categoria": categoria
    }
    datos["gastos"].append(nuevo_gasto)
    guardar_datos(datos)
    print(" Gasto registrado correctamente")

def buscar_gasto(id_gasto):
    datos = cargar_datos()
    for gasto in datos["gastos"]:
        if gasto["id"] == id_gasto:
            print(" Detalle del gasto:", gasto)
            return
    print(" Gasto no encontrado")

def actualizar_gasto(id_gasto, nueva_descripcion, nuevo_monto):
    datos = cargar_datos()
    saldo = calcular_balance()

    for gasto in datos["gastos"]:
        if gasto["id"] == id_gasto:
            diferencia = nuevo_monto - gasto["monto"]
            if diferencia > balance:
                print("Error: No tienes suficiente saldo para actualizar este gasto.")
                return
            gasto["descripcion"] = nueva_descripcion
            gasto["monto"] = nuevo_monto
            guardar_datos(datos)
            print("Gasto actualizado correctamente")
            return
    print(" Gasto no encontrado")

def eliminar_gasto(id_gasto):
    datos = cargar_datos()
    datos["gastos"] = [g for g in datos["gastos"] if g["id"] != id_gasto]
    guardar_datos(datos)
    print(" Gasto eliminado correctamente")

def registrar_ingreso(descripcion, monto, fecha):
    datos = cargar_datos()
    if monto <= 0:
        print(" Error: El monto del ingreso debe ser mayor que 0")
        return
    try:
        datetime.strptime(fecha, "%d/%m/%Y")
    except ValueError:
        print(" Error: Fecha inválida. Use formato dd/mm/aaaa")
        return
    nuevo_id = len(datos["ingresos"]) + 1
    nuevo_ingreso = {
        "id": nuevo_id,
        "descripcion": descripcion,
        "monto": monto,
        "fecha": fecha
    }
    datos["ingresos"].append(nuevo_ingreso)
    guardar_datos(datos)
    print("Ingreso registrado correctamente")
```

```python
%%writefile ui/consola.py
from core.gestor import (
    listar_gastos,
    crear_gasto,
    buscar_gasto,
    actualizar_gasto,
    eliminar_gasto,
    registrar_ingreso,
    calcular_balance
)

def menu():
    while True:
        print(" MENÚ PRINCIPAL")
        print("1. Listar gastos")
        print("2. Registrar gasto")
        print("3. Consultar gasto")
        print("4. Actualizar gasto")
        print("5. Eliminar gasto")
        print("6. Registrar ingreso")
        print("7. Ver balance actual")
        print("8. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            listar_gastos()

        elif opcion == "2":
            descripcion = input("Descripción del gasto: ")
            try:
                monto = float(input("Monto: "))
            except ValueError:
                print(" Error: El monto debe ser numérico")
                continue
            fecha = input("Fecha (dd/mm/aaaa): ")
            categoria = input("Categoría: ")
            crear_gasto(descripcion, monto, fecha, categoria)

        elif opcion == "3":
            try:
                id_gasto = int(input("ID del gasto a consultar: "))
                buscar_gasto(id_gasto)
            except ValueError:
                print("Error: El ID debe ser numérico")

        elif opcion == "4":
            try:
                id_gasto = int(input("ID del gasto a actualizar: "))
                nueva_descripcion = input("Nueva descripción: ")
                nuevo_monto = float(input("Nuevo monto: "))
                actualizar_gasto(id_gasto, nueva_descripcion, nuevo_monto)
            except ValueError:
                print(" Error: El ID y el monto deben ser numéricos")

        elif opcion == "5":
            try:
                id_gasto = int(input("ID del gasto a eliminar: "))
                eliminar_gasto(id_gasto)
            except ValueError:
                print(" Error: El ID debe ser numérico")

        elif opcion == "6":
            descripcion = input("Descripción del ingreso: ")
            try:
                monto = float(input("Monto: "))
            except ValueError:
                print(" Error: El monto debe ser numérico")
                continue
            fecha = input("Fecha (dd/mm/aaaa): ")
            registrar_ingreso(descripcion, monto, fecha)

        elif opcion == "7":
            balance = calcular_balance()
            print(f" Balance actual disponible: {balance}")

        elif opcion == "8":
            print(" Saliendo del sistema...")
            break

        else:
            print(" Opción inválida. Intente nuevamente.")
```
